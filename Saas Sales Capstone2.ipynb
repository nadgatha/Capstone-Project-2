## **1. DATA UNDERSTANDING**

### **A. Background Data**

Amazon Web Services (AWS) adalah sebuah penyedia layanan cloud computing yang didirikan oleh Amazon.com pada tahun 2006. AWS menyediakan berbagai layanan komputasi, salah satunya adalah SaaS yaitu "Software as a Service". Saas merupakan salah satu model layanan cloud computing di mana perangkat lunak atau aplikasi disediakan kepada pengguna melalui internet sebagai layanan berlangganan.

Segmentasi penjualan AWS SaaS dibagi menjadi 3 berdasarkan karakteristik tertentu, seperti profil pelanggan, preferensi pembelian, atau performa transaksi. Hal ini dapat membantu dalam pemahaman lebih mendalam tentang perilaku pelanggan, meningkatkan strategi pemasaran, dan mengoptimalkan hasil penjualan. Segmentasi penjualan AWS ialah: 
1. **"SMB"** atau Small and Medium-sized Businesses, yang dalam bahasa Indonesia dikenal sebagai "Usaha Kecil dan Menengah" (UKM).
2. **"Strategic"**, yaitu segmen yang tidak hanya didasarkan pada karakteristik umum seperti ukuran bisnis (SMB, perusahaan besar) atau geografi, tetapi juga mencerminkan strategi bisnis yang lebih dalam.
3. **"Enterprise"** adalah dalam konteks bisnis dan teknologi merujuk pada segmen pasar yang terdiri dari perusahaan-perusahaan besar atau korporasi.

Berdasarkan segmentasi penjualan AWS SaaS, maka dapat dilakukan analisis untuk mengetahui tren penjualan dan keuntungan dari industri AWS di tiap-tiap wilayah  untuk mencapai Goals yang dibutuhkan **Manajemen AWS** selaku stakeholder.

Setelah melihat tren profit yang dihasilkan AWS, maka disini kita perlu melihat apakah terdapat perbedaan profit secara signifikan pada tiap bulan dan tahun. Dari hasil tes, dapat dilihat bahwa tidak ada perbedaan profit secara signifikan dalam per bulan dan tahun walaupun profit yang diperoleh beragam.

```python
from scipy.stats import kruskal

# Membuat daftar tahun yang ada dalam data
years = profitTrend['Order Year'].unique()

# Menginisialisasi daftar kelompok data berdasarkan tahun
data_groups = []

# Memisahkan data penjualan ke dalam kelompok berdasarkan tahun
for year in years:
    data_year = profitTrend[profitTrend['Order Year'] == year]['Profit']
    data_groups.append(data_year)

# Melakukan uji Kruskal-Wallis
statistic, p_value = kruskal(*data_groups)

# Menampilkan hasil uji
print("Uji Kruskal-Wallis untuk Penjualan per Bulan dan Tahun:")
print("Nilai Statistik Kruskal-Wallis:", statistic)
print("Nilai p-value:", p_value)

# Menentukan apakah hasil uji signifikan
alpha = 0.05
if p_value < alpha:
    print("Terdapat perbedaan keuntungan secara signifikan dalam per bulan dan tahun.")
else:
    print("Tidak ada perbedaan keuntungan secara signifikan dalam per bulan dan tahun.")
```
